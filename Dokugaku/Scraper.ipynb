```python
import urllib.request
from bs4 import BeautifulSoup


class Scraper:
   
    def __init__(self, site):
        self.site = site

    def scrape(self):
        r = urllib.request.urlopen(self.site)
        html = r.read()
        parser = "html.parser"
        sp = BeautifulSoup(html,
                           parser)
        
        for tag in sp.find_all("a"):
            url = tag.get("href")
            if url is None:
                print("false")
                continue
            if "html" in url:
                print("\n" + url)
                print("true")
           

news = "https://yahoo.com/"
Scraper(news).scrape()
print("finish")
```
